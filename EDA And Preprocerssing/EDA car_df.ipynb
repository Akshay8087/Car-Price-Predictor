### 📊 Price Distribution by Company

The boxplot below visualizes the **distribution of car prices across different companies** in the dataset.

#### 🔍 What the Plot Shows:
- **Each box** represents the interquartile range (IQR) of car prices for a company — the middle 50% of the data.
- The **line inside each box** is the **median price**.
- **Whiskers** extend to show the range (excluding outliers).
- **Dots outside the whiskers** are **outliers**, indicating cars that are priced unusually high or low for that brand.

#### 🧠 Key Observations:
- **Luxury brands like Audi, Mercedes, BMW, Jaguar, and Land Rover** have a **much higher median price** and a wider price spread.
- **Companies like Maruti, Hyundai, Tata, and Honda** show **lower price ranges**, suggesting they offer more budget-friendly vehicles.
- Some brands (e.g., **Datsun, Mini, and Mitsubishi**) show **very narrow IQRs**, which may indicate fewer models or more consistent pricing.
- **Jaguar and Land Rover** have **consistently high prices** with fewer outliers, highlighting a premium pricing strategy.

#### 📌 Insights for Analysis:
- Understanding how **price varies by brand** can help in clustering vehicles, pricing models, or even in customer segmentation.
- You might consider using price as a **feature** in machine learning models like regression or classification.


#### Checking relationship of Year with Price


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size and style
plt.figure(figsize=(20, 10))
sns.set(style="whitegrid")

# Sort the dataframe by year for better x-axis order
car_sorted = car_df.sort_values(by='year')

# Create the swarm plot
ax = sns.swarmplot(x='year', y='Price', data=car_sorted)

# Customize the plot
ax.set_title("Price Distribution by Year", fontsize=18, fontweight='bold')
ax.set_xlabel("Manufacturing Year", fontsize=14)
ax.set_ylabel("Car Price (in ₹)", fontsize=14)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 📊 Price Distribution by Year

This visualization displays the distribution of **car prices** across different **manufacturing years** using a **swarm plot**.


#### ✅ Insights:

- Each dot represents the price of an individual car for a given manufacturing year.
- **X-axis:** Manufacturing Year  
- **Y-axis:** Car Price (in ₹)
- The plot reveals that cars manufactured between **2010 and 2017** have the **highest listing density**, possibly indicating greater availability or popularity during those years.
- Several **outliers** and **premium-priced vehicles** are noticeable, especially in the 2012–2016 range.
- Years prior to 2005 show relatively fewer listings, likely due to the aging and scarcity of older models.

#### 📌 Why Swarm Plot?

- Swarm plots display **all individual data points** without overlap, making it easier to observe **distribution and clusters**.
- Unlike boxplots, which summarize data, swarm plots retain full granularity—ideal for visualizing detailed price trends over time.



#### 💡 Suggestion:

If the chart appears cluttered due to large data volume, consider:
- Filtering for recent years (e.g., 2010 onward)
- Using a **violin plot** or **boxen plot** for smoother visual distribution

#### Relationship of Price with FuelType, Year and Company mixed


```python


# Create the relplot with enhanced aesthetics
ax = sns.relplot(x='company', y='Price', data=car_df, 
                 hue='fuel_type', size='year', 
                 height=7, aspect=2, 
                 palette='Set1', sizes=(30, 200))

ax.set_xticklabels(ax.ax.get_xticklabels(), rotation=40, ha='right')

ax.set(title="Price vs Company by Fuel Type and Year", 
       xlabel="Company", ylabel="Price")

ax.legend.set_title('Fuel Type')
ax.legend.set_bbox_to_anchor((1.05, 1))  # Places legend outside the plot
ax.legend.set_loc('upper left')
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(14, 6))

# Histogram + KDE
sns.histplot(car_df['kms_driven'], kde=True, color='mediumseagreen', edgecolor='black')
plt.title('Distribution of Kilometers Driven', fontsize=14, fontweight='bold')
plt.xlabel('Kilometers Driven')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.5)



plt.tight_layout()
plt.show()
```

```python


# Calculate kms_per_year (you already did this, just keeping for reference)
car_df['kms_per_year'] = car_df['kms_driven'] / car_df['car_age'].replace(0, 1)

plt.figure(figsize=(10, 6))
sns.boxplot(x='fuel_type', y='kms_per_year', data=car_df, palette='coolwarm')

plt.title('Annual Kilometers Driven by Fuel Type', fontsize=16, fontweight='bold')
plt.xlabel('Fuel Type', fontsize=12)
plt.ylabel('Kilometers Driven Per Year', fontsize=12)
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
